```python
# SVM : Support vector machine : It can handle multiple continous and categorical variables. 
# It creates a hyperplane and the values near the hyperplane are taken as support vectors. 
# Advanced topics in ML include SVM and Decision trees
# We need better accuracy for any method we use(Decision trees,KNN,SVM)
```

```python
import pandas as pd
ds = pd.read_csv("bill_authentication.csv")
ds.head(10)
```

```python
x = ds.drop('Class',axis=1)
y = ds['Class']
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(x,y, test_size = 0.4, random_state = 10)
from sklearn.svm import SVC
sv = SVC(kernel='linear')
sv.fit(x_train, y_train)
y_pred = sv.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
