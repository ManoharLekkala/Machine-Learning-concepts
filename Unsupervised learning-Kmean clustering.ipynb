```python
# labelled data is called supervised learning and opposite is unsupervised learning
# we make prediction using categorical data using supervised learning
# and we categorise the data it is unsupervised learning
```

```python
# we are importing dataset from sklearn datasets by importing make_blobs
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plot
data = make_blobs(n_samples= 150, n_features=2, centers=4,cluster_std=1.5,random_state=30)

# creating an numpy array for data points
p = data[0]
kmeans = KMeans(n_clusters=4)
# now fit kmeans object to data
kmeans.fit(p)
# print location of clusters based on kmeans object
print(kmeans.cluster_centers_)

# saving new clusters for chart
y= kmeans.fit_predict(p)
plot.scatter(p[y==0,0],p[y==0,1], s=200, c='black')
plot.scatter(p[y==1,0],p[y==1,1], s=200, c='blue')
plot.scatter(p[y==2,0],p[y==2,1], s=200, c='red')
plot.scatter(p[y==3,0],p[y==3,1], s=200, c='green')
```
